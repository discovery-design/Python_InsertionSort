```python
import requests as req 
import pandas as pd
import json
import matplotlib.pyplot as plt
jsondata=req.get("https://data.ibb.gov.tr/datastore/odata3.0/d5fe41b0-3848-4548-9ac7-6e4756c3027b?$format=json").text

data=json.loads(jsondata)
data1=pd.DataFrame(data['value'])

##Kadıköy datasını analiz edelim.
KData=data1[data1['Ilce']=='KADIKÖY']

KData.plot.bar()
##Horizontal
KData.plot.barh()

```
